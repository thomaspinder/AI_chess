```python
import chess
import chess.pgn
import numpy as np

pgn = open('data/2015_05.pgn')

no_games = 1000000
i = 0
move_counter = []
while i < no_games:
    if i*100%no_games == 0:
        print('{}% complete'.format(i*10/no_games))
    game = chess.pgn.read_game(pgn)
    move_counter.append(len(list(game.main_line())))
    i += 1

mean_moves = np.mean(move_counter)
median_moves = np.median(move_counter)
```

```python
import matplotlib.pyplot as plt
from collections import Counter, OrderedDict
%matplotlib inline
move_count = Counter(move_counter)
x, y =[], []
moves_soted = OrderedDict(move_count)

for k,v in sorted(move_count.items()):
    # The quickest possible mate (fools mate) is 4 moves
    if k > 4:
        x.append(k)
        y.append(v)

def roundup(x):
    return int(np.ceil(x / 10.0)) * 10

xlimit = roundup(np.max(x))
plt.figure(figsize=(10,6))
plt.plot(x, y, '--')
plt.xticks(np.arange(start=0, stop= xlimit, step = 25))
plt.xlabel('Move Count')
plt.ylabel('Occcurences')
plt.savefig('move_counts.png')
```
